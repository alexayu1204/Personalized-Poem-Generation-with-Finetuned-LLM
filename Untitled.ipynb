```python
from datasets import load_dataset,DatasetDict

ds = load_dataset("larryvrh/Chinese-Poems")
```

```python
# `clean up
ds = ds.map(lambda x: {'text': x['title'].strip() + '\n<|end|>\n' + x['content'].strip()})
ds = ds.remove_columns(['dynasty', 'author','title', 'content'])
```

```python
print(ds)
```

```python
train_testvalid = ds['train'].train_test_split(test_size=0.16)
# Split the 10% test + valid in half test, half valid
test_valid = train_testvalid['test'].train_test_split(test_size=0.1)
# gather everyone if you want to have a single DatasetDict
ds = DatasetDict({
    'train': train_testvalid['train'],
    'test': test_valid['test'],
    'valid': test_valid['train']})
```

```python
# Save in JSON Lines format
for split, dataset in ds.items():
    dataset.to_json(f"{split}.jsonl")
```
